```python
import folium
from folium.plugins import HeatMap
from folium import plugins
import pandas as pd
```

```python
dados = pd.read_excel('municipios.xlsx')
```

```python
dados.head()
```

```python
# Mapa Brasil
baseMap = folium.Map(
    width = '100%',
    height = '100%',
    location =[-15.788497, -47.899873],
    zoom_start = 4)
```

```python
baseMap
```

```python
coordenadas = dados[["LATITUDE", "LONGITUDE", "POPULACAO"]].values.tolist()
```

```python
coordenadas
```

```python
HeatMap(coordenadas, radius = 15).add_to(baseMap)
```

```python
baseMap
```

## Informações Tooltip

```python
dados.iloc[0]['LATITUDE']
```

```python
# Criando o cirulo tooltip
for i in range(0, len(dados)):
    folium.Circle(
        location=[dados.iloc[i]['LATITUDE'], dados.iloc[i]['LONGITUDE']],
        color='#000000',
        fill='#00A1B3',
        tooltip='<ul>' +
               '<li><b>Município:</b> ' + str(dados.iloc[i]['MUNICIPIO']) + '</li>' +
               '<li><b>Estado:</b> ' + str(dados.iloc[i]['UF']) + '</li>' +
               '<li><b>População:</b> ' + f'{int(dados.iloc[i]["POPULACAO"]):,}'.replace(',', '.') + '</li>' +
               '</ul>',
        radius=10
    ).add_to(baseMap)
```
